```python
def move_runners(situation, hit):
    runners = situation[0:3]
    sit = situation[-2:]
    if hit == 4:
        on_base = [bool(runner > 0) for runner in runners]
        for i in np.arange(len(runners)-1):
            if on_base[i]:
                runners[i] += 4
        runs = sum([bool(runner >= 4) for runner in runners]) + 1
        sit[1] = runs
        runners = [0, 0, 0]
        return runners + sit
    if sum(runners) == 0:
        runners[hit - 1] = hit
        return runners + sit
    else:
        on_base = [bool(runner > 0) for runner in runners]
        for i in np.arange(3):
            if on_base[2 - i]:
                move = runners[2 - i] + hit
                if move >= 4:
                    runners[2 - i] = 0
                    sit[1] += 1
                else:
                    runners[move - 1] = runners[2 - i] + hit
                    runners[2 - i] = 0
        runners[hit - 1] = hit
        return runners + sit
```
